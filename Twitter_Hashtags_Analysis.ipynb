```python
#This program processes the tweets scraped from the twitter search results for nine ICO campaigns, and gets all the hashtags 
#available in these tweets.
#The tweets to be processed are available in file "Tweet_Text" generated by program "icoTweetScrape.ipynb".
#The results are stored in a file named "hashtag_counts.csv" and holds the hashtags, and the corresponding counts for each 
```

```python
import pandas as pd
import re #regex
from collections import Counter
import csv
```

```python
tweetTextdf = pd.read_csv("Twitter.csv")
tweetTextdf.head()
```

```python
hashtagList = []

#Iterate through each tweet and grab the hashtags
for index, row in tweetTextdf.iterrows():
    #print( re.findall(r"#(\w+)", row["tweet"]))
    hashtags =  re.findall(r"#(\w+)", row["tweet"])
    
    #if the hastags list returned from above is not empty, add each individual hashtg to the list of previously accumulated 
    #hashtags
    if hashtags:
        for hashtag in hashtags:
            hashtagList.append( hashtag)

print(hashtagList)                      
```

```python
#Calculate the frequency of each hashtag to find th most popular ones
hashtagCounterDict = Counter(hashtagList)
```

```python
#Set up the file to write the hashtags and counts to. Note, parameter newline = '' helps avoid writing blank lines in between  
writefile = open('Twitter_Hashtags_Count.csv','w', newline='')
writer = csv.writer(writefile)

#write the hashtags alsong with corresponding count to a csv file "hashtag_counts.csv"
for key, count in hashtagCounterDict.items():
    writer.writerow([key, count])    

#Make sure to close your file to guarantee it gets flushed
writefile.close() 
```
